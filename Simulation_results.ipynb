```python
import opencor as oc
simulation = oc.open_simulation('total.cellml')
```

```python
data = simulation.data()
data.set_starting_point(0)
data.set_ending_point(3)
data.set_point_interval(0.001)
```

```python
simulation.run()
```

```python
results = simulation.results()
```

```python
ds=results.dataStore()
variables = ds.voi_and_variables()
variables
```

```python
Aortic_v=variables['block9/block8/block7/block6/block5/block4/block3/block2/block1/Aortic_root/AorticRoot/v_I'].values()  # Numpy array
Arch_II_v=variables['block9/block8/block7/block6/block5/block4/Arch_II/ArchII/v_I'].values()
Arch_I_v=variables['block9/block8/block7/block6/block5/block4/block3/block2/Arch_I/ArchI/v_I'].values()
Ascending_v=variables['block9/block8/block7/block6/block5/block4/block3/block2/block1/AscendingAorta/Ascending/v_I'].values()
leftCarotid_v=variables['block9/block8/block7/block6/block5/block4/block3/Left_common_carotid/LeftCarotid/v_I'].values()
leftSubClavian_v=variables['block9/block8/block7/block6/block5/Left_subclavian/LeftSubclavian/v_I'].values()
Descending_v=variables['block9/block8/block7/block6/Descending_Aorta/Descending/v_I'].values()
Brachiocephalic_v=variables['block9/block8/block7/Brachiocephalic_artery/Brachiocephalic/v_I'].values()
rightCarotid_v=variables['block9/block8/RightCarotid/Right_common_carotid/v_I'].values()
rightSubClavian_v=variables['block9/RightSubclavian/Right_subclavian/v_I'].values()

time=variables['LV/LV_output/t'].values()
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors
from pylab import rcParams
rcParams['figure.figsize'] = 10, 6
for c, flow, label in [('b', Aortic_v, 'Aortic root'),('c', Arch_II_v, 'Arch of Aorta II'), 
                       ('m', Arch_I_v, 'Arch of Aorta I'),('r', leftCarotid_v, 'Left common Carotid'),
                       ('y', leftSubClavian_v, 'Left Sub Clavian'),('grey', Descending_v, 'Descending Aorta'),
                       ('violet', Brachiocephalic_v, 'Brachiocephalic'), ('lightseagreen', rightCarotid_v, 'Right common Carotid'),
                       ('chartreuse', rightSubClavian_v, 'Right Sub Clavian')]:
    plt.plot(time,flow, c, label=label)
    plt.xlabel("time (s)")
    plt.title('Blood flow (m3/s)')
    plt.legend(loc='upper right')

plt.grid()
```
